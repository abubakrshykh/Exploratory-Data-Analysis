```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
feature_names = ['sepal_length','sepal_width','petal_length','petal_width']
```

```python
data = pd.DataFrame(data = iris.data , columns = feature_names)
data.head()
```

```python
data['target'] = iris.target
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(data['sepal_length'], bins = 35)

plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
#SCATTER PLOT

plt.scatter(data['sepal_length'], data['sepal_width'], c=data['target'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

plt.scatter(data['petal_length'], data['petal_width'], c=data['target'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
#line plot
plt.plot(data['sepal_length'])
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.show()

plt.plot(data['sepal_width'])
plt.xlabel('Index')
plt.ylabel('Sepal Width')
plt.show()

plt.plot(data['petal_length'])
plt.xlabel('Index')
plt.ylabel('petal Length')
plt.show()

plt.plot(data['petal_width'])
plt.xlabel('Index')
plt.ylabel('petal Width')
plt.show()
```

```python
#BAR CHART
species_count = data['target'].value_counts()
plt.bar(species_count.index, species_count.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
data['target'].value_counts().index
```

```python
#HEATMAP
corr = data.corr()
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.show()
```

```python
data2 = data
data2= data.drop(['sepal_width'], axis=1)
```

```python
data2
```

```python
#APPLYING BINNING TECHNIQUE


# Discretize petal length into three bins
data2['petal_length_bin'] = pd.cut(data2['petal_length'], bins=3, labels=['short', 'medium', 'long'])

# Print the first 10 rows of the binned iris data
data2
```

```python
col = data2.pop('target')
data2.insert(len(data2.columns), 'target', col)
```

```python
data2
```

```python
data2['petal_length_bin_categories'] = data2.petal_length_bin.map({'short':0, 'medium':1, 'long':2})
data2 = data2.drop(['petal_length_bin'],axis=1)
```

```python
data2
```

```python
X=data2.iloc[:,:-1].values
y=data2.iloc[::,-1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
accuracy =model.score(X_test, y_test)
print("Model's Accuracy is: ", accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression(max_iter=1000)
model2.fit(X_train,y_train)
```

```python
accuracy =model2.score(X_test, y_test)
print("Model's Accuracy is: ", accuracy)
```
